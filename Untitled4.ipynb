```python
def feature_selection(df):
    df["hour"]=df.index.hour
    df["day_of_week"]=df.index.dayofweek
    df["quarter"]=df.index.quarter
    df["month"]=df.index.month
    df["year"]=df.index.year
    df["day_of_year"]=df.index.dayofyear
    return df 

ev=feature_selection(ev)
test=ev[ev.index>="01-01-2015"]
train=ev[ev.index<"01-01-2015"]
```

```python
fig=plt.figure(figsize=(16,9))
ax=fig.add_axes([0,0,.8,.8])
ax.plot(train,color="blue",ls="-",marker="o",label="Train", mfc = 'r')
ax.plot(test,color="green",ls="-",marker="o",label="Test", mfc = 'y')
plt.title("Train / Test --- Split")
plt.xlabel("Date")
plt.ylabel("Electric_Production")
```

```python
x_train=train.drop("AEP_MW",axis=1)
y_train=train[["AEP_MW"]]

x_test=test.drop("AEP_MW",axis=1)
y_test=test[["AEP_MW"]]
```

```python
xgb_reg=xgb.XGBRegressor(n_estimators=1000,early_stopping_rounds=50,learning_rate=0.01)
```

```python
xgb_reg.fit(x_train,y_train,eval_set=[(x_train,y_train),(x_test,y_test)], verbose=100)
```

```python
y_pred=xgb_reg.predict(x_test)
```

```python
fig=plt.figure(figsize=(16,9))
plt.plot(ev.index,ev["AEP_MW"])
plt.plot(x_test.index,y_pred)
```

```python
score=np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
import pandas as pd
from datetime import timedelta, datetime
import matplotlib.pyplot as plt


last_date = ev.index[-1]


days_to_generate = 365
hours_to_generate = days_to_generate * 24  

all_dates = [last_date + timedelta(hours=i) for i in range(1, hours_to_generate + 1)]

# Create DataFrame from future dates
df = pd.DataFrame(all_dates, columns=["date"])
df = df.set_index("date")


def feature_selection(df):
    df["hour"] = df.index.hour
    df["day_of_week"] = df.index.dayofweek
    df["quarter"] = df.index.quarter
    df["month"] = df.index.month
    df["year"] = df.index.year
    df["day_of_year"] = df.index.dayofyear
    return df

df = feature_selection(df)

# Fit the XGBoost model on the existing data (X, Y)
X = ev.drop("AEP_MW", axis=1)
Y = ev["AEP_MW"]

xgb_reg = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.01)
xgb_reg.fit(X, Y, verbose=100)


y_pred = xgb_reg.predict(df)


fig = plt.figure(figsize=(16, 9))
plt.plot(ev.index, ev["AEP_MW"], label="Actual Data")   # Plot existing data
plt.plot(df.index, y_pred, label="Predicted Future", color="orange")  # Plot future predictions
plt.legend()
plt.show()

```
